### Q1. Ask the user to enter a value (e.g., M) for a mathematical function. 
- Generate x values from -10 to 10 using np.linspace().
- Compute y values for:
    - y= M⋅x^2
    - y=M⋅sin(x)


### Plot both functions in a single figure:
- Use different colors and line styles.
- Add legend, grid, and title.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
m=eval(input('Enter value of m: '))
x=np.linspace(-10,10,500)
y1=m*x**2
y2=m*np.sin(x)
plt.plot(x,y1,'r--',label='y=m*x^2')
plt.plot(x,y2,'b-',label='y=m*sin(x)')
plt.legend()
plt.grid()
plt.title('Mathematical functions')
plt.xlabel('x -->')
plt.ylabel('y -->')
plt.show()
```

### Q2. Create a dataset of five subjects and scores.
- Convert it into a Pandas DataFrame.
- Plot the scores using a Seaborn bar plot with:
    - Different colors for each bar.
    - Annotations on top of each bar.
    - Title, axis labels, and grid.

```python
dict={
    'Subject':['Maths','Physics','Typography','English','Chemistry'],
    'Score':[97,95,100,88,95]
}
df=pd.DataFrame(dict)
plt.figure(figsize=(10,5))
sns.barplot(x='Subject', y='Score', data=df, hue='Subject', palette='rainbow', legend=False)
for index, value in enumerate(df['Score']):
    plt.text(index, value, f'{value}', ha='center', fontsize=12)
plt.title('Scores by Subject', fontsize=14)
plt.xlabel('Subjects -->', fontsize=12)
plt.ylabel('Scores -->', fontsize=12)
plt.xticks(rotation=10)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
```

### Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().
- Create a 2x2 subplot layout displaying:
    - Line plot showing cumulative sum.
    - Scatter plot with random noise.
- Customize with titles, axis labels, and grids.

```python
roll_number =102317174
np.random.seed(roll_number)
data = np.random.randn(50)
fig, axes = plt.subplots(1, 2, figsize=(10,5))
axes[0].plot(np.cumsum(data), label='Cumulative Sum', color='blue',alpha=0.7)
axes[0].set_title('Line plot of Cumulative Sum', fontsize=14)
axes[0].set_xlabel('Index -->')
axes[0].set_ylabel('Cumulative Sum  -->')
axes[0].grid(True)
axes[1].scatter(range(len(data)), data, color='red', label='Random Noise',alpha=0.8)
axes[1].set_title('Scatter Plot with Random Noise', fontsize=14)
axes[1].set_xlabel('Index -->')
axes[1].set_ylabel('Value -->')
axes[1].grid(True) 
plt.tight_layout()
plt.show()
```

### Q4. Download Data-set from the below link https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv. Apply ‘seaborn’ library to do the following.
- Read Total profit of all months and show it using a line plot.
- Read all product sales data and show it using a multiline plot.
- Plot bar chart for all the features/attributes.

```python
url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='month_number', y='total_profit', data=df, marker='o')
plt.title('Total Profit of All Months', fontsize=15)
plt.xlabel('Month Number -->')
plt.ylabel('Total Profit -->')
plt.grid(True)
plt.show()
```

```python
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
plt.figure(figsize=(12, 8))
for product in product_columns:
    sns.lineplot(data=df, x='month_number', y=product, marker='o', label=product)
plt.title('Monthly Sales Data for Each Product')
plt.xlabel('Month Number -->')
plt.ylabel('Sales Units -->')
plt.legend(title='Products')
plt.grid(True)
plt.show()
```

```python
df_products = df.drop(columns=['month_number', 'total_units', 'total_profit'])
total_sales = df_products.sum().reset_index()
total_sales.columns = ['Product', 'Total Sales']
plt.figure(figsize=(10,5))
sns.barplot(data=total_sales, x='Product', y='Total Sales',hue='Product', palette='rainbow')
plt.title('Total Sales for Each Product')
plt.xlabel('Products -->')
plt.ylabel('Total Sales -->')
plt.xticks(rotation=45) 
plt.grid(axis='y')
plt.show()
```
